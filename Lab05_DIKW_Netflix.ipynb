```python
#-----2. Limpieza de datos
#Identificar valores nulos
df.isnull().sum()
```

```python
#Calcular porcentajes de valores faltantes
porcentaje_nulos = df.isnull().mean() *100
```

```python
#Detectar registros duplicados
df.duplicated().sum()
```

```python
#Aplicar estrategias de limpieza, eliminacion y tranformacion
df_limpio = df.copy()
```

```python
df_limpio["director"] = df_limpio["director"].fillna("No especificado")
```

```python
df_limpio["cast"] = df_limpio["cast"].fillna("No especificado")
```

```python
df_limpio["country"] = df_limpio["country"].fillna("No especificado")
```

```python
df_limpio["rating"] = df_limpio["rating"].fillna("No especificado")
```

```python
df_limpio["duration"] = df_limpio["duration"].fillna("No especificado")
```

```python
df_limpio["date_added"] = df_limpio["date_added"].fillna("No especificado")
```

```python
df_limpio = df_limpio.drop_duplicates()
```

```python
df_limpio.isnull().sum()
```

```python
#Justificar las deciciones tomadas
#Se decidió reemplazar los valores nulos por "No especificado" 
#porque columnas como director, elenco, país o fecha agregada son datos 
#descriptivos importantes. Eliminarlos podría reducir demasiado la 
#información del dataset. También se eliminaron registros duplicados para 
#evitar conteos repetidos en el análisis.
```

```python
#-----3. Analisis descriptivo(Capa de informacion)
plt.figure(figsize=(6,4))
sns.countplot(data=df_limpio, x="type")
plt.title("Cantidad de contenido por tipo")
plt.xlabel("Tipo de contenido")
plt.ylabel("Cantidad")
plt.savefig("grafica_1_tipo_contenido.png")
plt.show()
```

```python
top_paises = df_limpio["country"].value_counts().head(10)

plt.figure(figsize=(10,5))
top_paises.plot(kind="bar")
plt.title("Top 10 países con más contenido en Netflix")
plt.xlabel("País")
plt.ylabel("Cantidad de títulos")
plt.xticks(rotation=45)
plt.savefig("grafica_2_top_paises.png")
plt.show()
```

```python
df_limpio["date_added"] = pd.to_datetime(df_limpio["date_added"], errors="coerce")
df_limpio["year_added"] = df_limpio["date_added"].dt.year

contenido_anio = df_limpio["year_added"].value_counts().sort_index()

plt.figure(figsize=(10,5))
contenido_anio.plot(kind="line", marker="o")
plt.title("Contenido agregado a Netflix por año")
plt.xlabel("Año")
plt.ylabel("Cantidad de títulos")
plt.savefig("grafica_3_contenido_por_anio.png")
plt.show()
```

```python
top_rating = df_limpio["rating"].value_counts().head(10)

plt.figure(figsize=(10,5))
top_rating.plot(kind="bar")
plt.title("Clasificaciones más comunes en Netflix")
plt.xlabel("Clasificación")
plt.ylabel("Cantidad")
plt.xticks(rotation=45)
plt.savefig("grafica_4_clasificaciones.png")
plt.show()
```

```python
#------4. Interpretacion de patrones (CAPA DE CONOCIMIENTOS)
#Escribe al menos tres concluciones derivadas del analisis

#1. Netflix cuenta con una mayor cantidad de películas que series, lo que indica que su catálogo se ha 
#enfocado principalmente en contenido de formato largo y único.

#2. Los países con mayor producción o presencia de contenido suelen ser Estados Unidos, India y 
#Reino Unido, lo cual muestra que Netflix concentra gran parte de su catálogo en mercados con alta 
#producción audiovisual.

#3. El crecimiento del contenido agregado por año permite observar en qué periodos Netflix 
#incrementó más su catálogo. Esto ayuda a identificar años clave de expansión de la plataforma.

#4. Las clasificaciones más frecuentes permiten entender hacia qué tipo de público está dirigido el 
#contenido. Si predominan clasificaciones para adolescentes o adultos, se puede concluir que Netflix 
#orienta gran parte de su catálogo a esos segmentos.
```
