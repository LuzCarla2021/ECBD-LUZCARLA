## Importacion de las librerias

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report,
    ConfusionMatrixDisplay
)
```

```python
##Carganmos el dataset Iris
from sklearn.datasets import load_iris
iris = load_iris(as_frame = True)
```

```python
dfIris = iris.frame
```

```python
dfIris.head()
```

## Exploracion del conjunto de datos


```python
dfIris.shape
```

```python
dfIris.columns
```

```python
iris.target_names
```

```python
## renombrar la columna target a los nombres de las especies
dfIris.rename(columns = {"target": "especie"}, inplace = True)
```

```python
dfIris["especie"] = dfIris["especie"].map({
    0: "setosa",
    1: "versicolor",
    2: "virginica"
})
```

```python
dfIris.head()
```

```python
dfIris.tail()
```

## Identificar variables predictoras (x) y las variables objetivo (y)

```python
# variables predictoras
X = dfIris.drop(columns = ["especie"])
```

```python
print("Variables predictoras")
print(X.columns)
```

```python
# variables objetivo
y = dfIris["especie"]
```

```python
print("Variables objetivo")
print(y.name)
```

## Dividir los datos en prueba y entrenamiento

80% para entrenamiento
20% para prueba

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.20, random_state = 42, stratify = y 
)
```

```python
print("X_train", X_train.shape)
print("X_test", X_test.shape)
print("y_train", y_train.shape)
print("y_test", y_test.shape)
```

## Escalado de variables

```python
scaler = StandardScaler()
```

```python
# escalamos las variables
X_train_scaler = scaler.fit_transform(X_train)
X_test_scaler = scaler.transform(X_test)
```

```python
print("sin escalar")
print(X_train.head())
print("escalado")
print(X_train_scaler[:5])
```

## Entrenamiento del modelo de KNN

En esta prueba utilizaremos el modelo de k = 3

```python
modelo = KNeighborsClassifier(n_neighbors = 3)
```

```python
modelo.fit(
    X_train_scaler, y_train
)
```

## Realizamos predicciones con las variables de prueba

```python
prediccion = modelo.predict(X_test_scaler)
```

```python
print(prediccion[:10])
```

```python
# Comparacion entre valores reales y valores predecibles
comparacion = pd.DataFrame({
    "especie real": y_test.values,
    "especie pre-dicha": prediccion
})
```

```python
comparacion.head(10)
```

## Evaluacion del modelo

```python
accuracy = accuracy_score(y_test, prediccion)
print (f"accuracy/exactitud: {accuracy: .4f}")
```

```python
# matriz de confucion -- igual en los modelos de la tarea

cm = confusion_matrix(y_test, prediccion)
```

```python
cm_df = pd.DataFrame(
    cm,
    index = modelo.classes_,
    columns = modelo.classes_
)
cm_df
```

## CLASIFICATION REPORT

PROPORCIONA METRICAS QUE PERMITE EVALUAR EL DESEMPEÑO DE CADA CLASE: PRECISION, RECALL, F1_SCORE AND SUPORT

```python
print(classification_report(y_test, prediccion))
```
