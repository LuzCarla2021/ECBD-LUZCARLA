```python
cantidad= 1000000
```

```python
import random
```

```python
nombres = ["Ana","luis","Martha","Carolina","Sofia","Pedro","Elena","Diego"]
ciudades = ["CDMX","Monterrey","Guadalajara","Puebla","Cancun","Merida","Tijuana","Veracruz"]
categorias = ["Tecnologia","Ropa","Hogar","videojuegos","Deportes","Calzado","Juguetes"]
```

```python
dfMillon = pd.DataFrame({
    "nombre" : [random.choice(nombres) for _ in range(cantidad)],
    "edad" : [random.randint(18, 70) for _ in range(cantidad)],
    "salario" : [random.randint(20000, 100000) for _ in range(cantidad)],
    "activo" : [random.choice([True, False]) for _ in range(cantidad)],
    "categoria" : [random.choice(categorias) for _ in range(cantidad)],
    "ciudad" : [random.choice(ciudades) for _ in range(cantidad)],
    "anio_registro" : [random.randint(2020, 2026) for _ in range(cantidad)]
})
```

```python
dfMillon.memory_usage(deep=True).sum() /1024**2
```
