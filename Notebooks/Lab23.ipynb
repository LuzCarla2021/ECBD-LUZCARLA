## Lab23: Evaluacion de modelos de regresion

### MAE, MSE, RMSE, R2

```python
# Importar librerias
import pandas  as pd
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.metrics import (
mean_absolute_error,
mean_squared_error,
root_mean_squared_error,
r2_score
)
```

```python
import matplotlib.pyplot as plt
```

```python
# importamos el dataset
diabetes = load_diabetes(as_frame = True)
print(diabetes.DESCR)
```

```python
df = diabetes.frame
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Seleccion de variables predictoras y variables objetivo

### Hipotesis: Existe una relacion entre las varibales clinicas del paciente y la propagacion un año despues

```python
X = df.drop(columns = "target")
```

```python
y = df["target"]
```

```python
X.head()
```

```python
y.head()
```

## Separacion de datos para entrenamiento y prueba

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    test_size = 0.3, 
    random_state = 42
)
```

## Entrenamiento

```python
modelo = LinearRegression()
modelo.fit(X_train, y_train)
```

## Predicciones

```python
prediccion = modelo.predict(X_test)
```

```python
comparacion = pd.DataFrame({
    "real": y_test.values,
    "prediccion": prediccion
}).reset_index(drop=True)

# Gráfica
plt.figure(figsize=(12, 6))

for i in range(len(comparacion)):
    plt.plot(
        [i, i],
        [comparacion.loc[i, "real"], comparacion.loc[i, "prediccion"]],
        color="grey",
        alpha=0.4
    )

plt.scatter(
    comparacion.index,
    comparacion["real"],
    label="Valor real",
    marker="o",
    s=70
)

plt.scatter(
    comparacion.index,
    comparacion["prediccion"],
    label="Predicción",
    marker="s",
    s=70
)

plt.xlabel("Paciente")
plt.ylabel("Valor del target")
plt.title("Valores reales vs. predichos")
plt.legend()
plt.grid(True)
plt.show()
```

## Metricas para finalizar

```python
# MAE
mae = mean_absolute_error(
    y_test,
    prediccion
)
```

```python
print(f"mae: {mae: .2f}")
```

```python
# MSE
mse = mean_squared_error(
    y_test,
    prediccion
)
```

```python
print(f"mse: {mse: .2f}")
```

```python
# RMSE
rmse = root_mean_squared_error(
    y_test,
    prediccion
)
```

```python
print(f"rmse: {rmse: .2f}")
```

```python
# Coeficiente de determinacion R2

r2 = r2_score(
    y_test,
    prediccion
)
```

```python
print(f"r2: {r2: .2f}")
```
